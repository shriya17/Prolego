```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ScoreBoardFinal.csv')
print df.head(5)
```

```python
df_copy = df.loc[df['Home_ShotsT'] != "XX-XX-"]
print df_copy
```

```python
#plt.figure(figsize=(10,10))

df_poss = df_copy[(df_copy['Home_Team']=="Arsenal") | (df_copy["Home_Team"] == "ARS")]
#print df_poss
df_poss = df_poss[['MId','Home_Poss','Away_Team']]
df_poss['index'] = range(1, len(df_poss) + 1)
df_poss[['MId','Home_Poss']] = df_poss[['MId','Home_Poss']].astype(float)
print df_poss
```

```python
plt.figure()
df_poss.set_index('Away_Team').plot(kind='bar',y='Home_Poss',figsize=(10,10))
plt.show()
```
